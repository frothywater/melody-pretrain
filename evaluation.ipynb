```python
import pandas as pd
```

```python
def get_mean_std(df: pd.DataFrame):
    groupby = df.groupby("metric").value
    return pd.DataFrame({"mean": groupby.mean(), "std": groupby.std()})
```

```python
def show_table(csv_path: str):
    data = pd.read_csv(csv_path)
    
    test_data = data.loc[data.group == "test", ["metric", "value"]]
    absolute_data = data[(data.kind == "mean") & (data.group != "test")].drop(columns=["kind"]).copy()
    relative_data = data[data.kind == "oa"].drop(columns=["kind"])
    relative_data.metric = "oa_" + relative_data.metric
    error_data = data[data.kind == "error"].drop(columns=["kind"])

    relative_data = get_mean_std(relative_data)
    error_data = get_mean_std(error_data)
    absolute_data = get_mean_std(absolute_data)
    df = pd.concat([relative_data, error_data, absolute_data]).reset_index()
    df["value"] = df["mean"].round(4).astype(str) + "±" + df["std"].round(4).astype(str)
    return df, test_data
```

```python
data, test_data = show_table("experiment/final/result/clm_epoch10.csv")
data
```
