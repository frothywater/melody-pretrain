```python
import pandas as pd
```

```python
def get_pivot(df: pd.DataFrame):
    df = df.groupby(["model", "task", "metric"])
    df = pd.DataFrame({"mean": df.value.mean()}).reset_index()
    df = df.pivot(index=["model"], columns="metric", values="mean")
    df = df.rename_axis(None, axis=0).rename_axis(None, axis=1)
    return df
```

```python
def show_table(csv_path: str):
    data = pd.read_csv(csv_path)
    
    test_data = data.loc[data.model == "test", ["metric", "value"]]
    absolute_data = data[(data.kind == "mean") & (data.model != "test")].drop(columns=["kind"]).copy()
    difference_data = absolute_data.merge(test_data, on=["metric"], suffixes=["", "_test"]).copy()
    difference_data["value"] = difference_data.value - difference_data.value_test
    difference_data = difference_data.drop(columns=["value_test"])

    relative_data = data[data.kind == "oa"].drop(columns=["kind"])
    error_data = data[data.kind == "error"].drop(columns=["kind"])

    relative_data = get_pivot(relative_data).add_prefix("OA_")
    error_data = get_pivot(error_data).rename(columns={"bar_pair_similarity": "structure_error"})
    absolute_data = get_pivot(difference_data)
    df = pd.concat([relative_data, error_data, absolute_data], axis=1).transpose()
    return df
```

```python
show_table("experiment/ablation_repr/result/metric_early.csv")
```
