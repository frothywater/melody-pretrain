```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("experiment/ablation_recovery/result/oa.csv")
data
```

```python
# absolute metrics
absolute_data = data[data.kind == "mean"].drop(columns=["kind"])
absolute_data.loc[absolute_data.model == "test", "sample"] = 0
absolute_data.loc[absolute_data.model == "test", "task"] = "clm"

absolute_data = absolute_data.groupby(["model", "task", "metric"])
absolute_data = pd.DataFrame({"mean": absolute_data.value.mean(), "std": absolute_data.value.std()}).reset_index()

absolute_data["mean"] = absolute_data["mean"].round(3).astype(str) + " ± " + absolute_data["std"].round(3).astype(str).replace("nan", "")
absolute_data = absolute_data.drop(columns=["std"])
absolute_data = absolute_data.pivot(index=["model", "task"], columns="metric", values="mean")
absolute_data
```

```python
# relative metrics
relative_data = data[data.kind == "oa"].drop(columns=["kind"])
relative_data = relative_data.groupby(["model", "task", "metric"])
mean = relative_data["value"].mean()
std = relative_data["value"].std()
relative_data = pd.DataFrame({"mean": mean, "std": std}).reset_index()

relative_data["mean"] = relative_data["mean"].round(3).astype(str) + " ± " + relative_data["std"].round(3).astype(str)
relative_data = relative_data.drop(columns=["std"])
relative_data = relative_data.pivot(index=["model", "task"], columns="metric", values="mean")
relative_data
```
