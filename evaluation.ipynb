```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("experiment/ablation_mask/result/metric.csv")
data
```

```python
# absolute metrics
absolute_data = data[data.kind == "mean"].drop(columns=["kind"])
absolute_data.loc[absolute_data.model == "test", "sample"] = 0
absolute_data.loc[absolute_data.model == "test", "task"] = "clm"

absolute_data = absolute_data.groupby(["model", "task", "metric"])
absolute_data = pd.DataFrame({"mean": absolute_data.value.mean()}).reset_index()
absolute_data = absolute_data.pivot(index=["model", "task"], columns="metric", values="mean")
absolute_data
```

```python
# relative metrics
relative_data = data[data.kind == "oa"].drop(columns=["kind"])

relative_data = relative_data.groupby(["model", "task", "metric"])
relative_data = pd.DataFrame({"mean": relative_data["value"].mean()}).reset_index()
relative_data = relative_data.pivot(index=["model", "task"], columns="metric", values="mean")
relative_data
```

```python
# error metrics
error_data = data[data.kind == "error"].drop(columns=["kind"])

error_data = error_data.groupby(["model", "task", "metric"])
error_data = pd.DataFrame({"mean": error_data["value"].mean()}).reset_index()
error_data = error_data.pivot(index=["model", "task"], columns="metric", values="mean")
error_data
```
